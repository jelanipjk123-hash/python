```python
import math

u = int(input("Enter no of values: "))

v = []   # square roots
w = []   # factorials

for i in range(u):
    x = int(input("Enter a value: "))

    y = math.sqrt(x)
    z = math.factorial(x)

    v.append(y)
    w.append(z)

print("Square roots:", v)
print("Factorials:", w)
```

```python
 u = int(input("Enter number of GR values: "))

gr_values = []

# Input GR values
for i in range(u):
    gr = float(input(f"Enter GR value {i+1} are: "))
    gr_values.append(gr)

# Calculate GR min and max
gr_min = min(gr_values)
gr_max = max(gr_values)

vsh_values = []

# Calculate Vsh
for gr in gr_values:
    vsh = (gr - gr_min) / (gr_max - gr_min)
    vsh_values.append(vsh)

print("\nGR Values:", gr_values)
print("Vsh Values:", vsh_values)
```
